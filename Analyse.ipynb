```python
import os
import librosa
import librosa.display
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import IPython.display as ipd
from tensorflow.python.platform import gfile
import tensorflow as tf
import numpy as np
import random

#for removing future warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Dataset
"""
PATH_TRAIN = r"E:/DATA/python/3.6/RNN-Ayukngaji/data/train"
```

```python
path = os.path.join(PATH_TRAIN,'*','*.wav')
wave_path = gfile.Glob(path)
```

```python
def graph(filename):
    wave, sr = librosa.load(filename, mono=True, duration=0.6)
    plt.figure(figsize=(6,4))
    librosa.display.waveplot(wave, sr=sr)

def mfcc(filename):
    wave, sr = librosa.load(filename, mono=True, duration=0.6)
    print(filename)
    mfccs = librosa.feature.mfcc(y=wave, sr=sr)
    print(mfccs.shape)
    #mfccs = np.pad(mfccs,((0,0), (0, 44 - len(mfccs[0]))), mode='constant')
    plt.figure(figsize=(6,4))
    plt.title(filename)
    librosa.display.specshow(mfccs,x_axis='time')
    plt.colorbar()
```

```python
for i in range(6,10):
    random.shuffle(wave_path)
    graph(wave_path[i])
    mfcc(wave_path[i])
    ipd.Audio(wave_path[i])
```
