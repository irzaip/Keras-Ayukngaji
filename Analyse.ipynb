```python
import os
import librosa
import librosa.display
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import IPython.display as ipd
from tensorflow.python.platform import gfile
import tensorflow as tf
import numpy as np
import random

#for removing future warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Dataset
"""
PATH_TRAIN = r"E:/tmp/speech_dataset"
```

```python
path = os.path.join(PATH_TRAIN,'*','*.wav')
wave_path = gfile.Glob(path)
```

```python
def graph(filename):
    wave, sr = librosa.load(filename, mono=True, duration=1)
    plt.figure(figsize=(6,4))
    librosa.display.waveplot(wave, sr=sr)

def mfcc(filename):
    wave, sr = librosa.load(filename, mono=True, duration=1)
    print(filename)
    mfccs = librosa.feature.mfcc(y=wave, sr=sr, n_mfcc=25)
    print(mfccs.shape)
    #mfccs = np.pad(mfccs,((0,0), (0, 44 - len(mfccs[0]))), mode='constant')
    plt.figure(figsize=(6,4))
    plt.title(filename)
    librosa.display.specshow(mfccs,x_axis='time')
    plt.colorbar()
    
def stft(filename):
    wave, sr = librosa.load(filename, mono=True, duration=1)
    print(filename)
    mfccs = librosa.feature.chroma_stft(wave,n_fft=8092)
    #mfccs = librosa.feature.mfcc(y=wave, sr=sr, n_mfcc=25)
    print(mfccs.shape)
    #mfccs = np.pad(mfccs,((0,0), (0, 44 - len(mfccs[0]))), mode='constant')

    plt.figure(figsize=(6,4))
    plt.title(filename)
    librosa.display.specshow(mfccs,x_axis='time')
    plt.colorbar()
    
def spec(filename):
    wave, sr = librosa.load(filename, mono=True)
    print(filename)
    D = librosa.amplitude_to_db(librosa.stft(wave))
    librosa.display.specshow(D, y_axis='log')
    #plt.colorbar(format='%+2.0f dB')
    print(D.shape)
    #plt.figure(figsize=(2,4))
    plt.title(filename)
    plt.colorbar()
```

```python
for i in range(100,115):
    #random.shuffle(wave_path)
    graph(wave_path[i])
    spec(wave_path[i])
```
