```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
import os
import random
import wave
import pyaudio
import scipy.io.wavfile as ww
from scipy.io.wavfile import read
```

```python
PATH_NOISE= os.getcwd()+'/data/noise'

chunk = 2048
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 16000
length_capture = 20  #20 ms
overlap_frame = 4    #4 ms
start_at = 7         #maybe not used
end_at = 100
DEBUG = False
```

```python
def getlistwav(path, dir=""):
    if dir == "": dir = path + "/"
    tmp = os.listdir(path)
    result = []
    for i in tmp:
        if str(i).find('wav',0)>0: result.append(dir+i)
    return result
```

```python
def writefile(filename, data):
    p = pyaudio.PyAudio()
    wf = wave.open(filename, 'wb')
    wf.setnchannels(CHANNELS)
    wf.setsampwidth(p.get_sample_size(FORMAT))
    wf.setframerate(RATE)
    wf.writeframes(data)
    wf.close() 

def addblank(filename, front=0, back=0):
    sr, wv = read(filename)
    wv=np.pad(wv,(front,back),'constant')
    plt.plot(wv)
    return wv

def getlen(filename):
    sr, wv = read(filename)
    print(len(wv))
    return len(wv)

def pitch(filename, target, n_steps=0.0):
    sr, wv = read(filename)
    a_max = wv.argmax()
    wv = wv / a_max
    y_ = librosa.effects.pitch_shift(wv, sr, n_steps=n_steps)
    wv = y_ * a_max
    wv = wv.astype('int16')
    writefile(target,wv)

def add_random_noise(filename, sr=22050, offset=0.0, duration=None):

    #add some blanks front and back
    front = random.randint(0,5000)
    back = random.randint(0,5000)

    
    wv=np.pad(wv,(front,back),'constant')
    
    #random pitch
    ptch = random.uniform(-3,3)
    wv = librosa.effects.pitch_shift(wv, sr, n_steps=ptch)
    
    #get the noise
    lnoise = getlistwav(PATH_NOISE, dir="",)
    random.shuffle(lnoise)
    wv2, sr2 = librosa.load(lnoise[0])
    rv = random.randint(0,len(wv2)-len(wv))
    wv2 = wv2[rv:rv+len(wv),]
    #wv3 = ((wv * random.uniform(0.9,1.3)) + (wv2 * random.uniform(0,0.4))) / 2
    wv3 = ( wv + wv2 ) / 2
    wv3 = np.asarray(wv3, dtype=np.int32)
    
    #ww.write('outa.wav',22050, wv3)    

```
